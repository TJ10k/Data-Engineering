##**Introduction:**
In Pyplot, you can use the **scatter()** function to draw a scatter plot.

The **scatter()** function plots one dot for each observation. It needs two arrays of the same length. One for the values of the x-axis, and one for values on the y-axis:

**Syntax:**
```
plt.scatter(x, y, color="red", s='', c='',cmap='',alpha='')

Hereâ€™s a summary of key points to remember about the main input parameters:

- x and y:	These parameters represent two variables we want to show the relationship.
- s: Defines the marker size.
- c: Represents the marker color.
- marker: Customizes the shape of the marker.
- cmap: Selects the mapping between values and colors.
- alpha: This parameter is a float number and represents the transparency of the markers.

```

##**Example 1: Creating Simple Scatter Plots**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.grid()
plt.scatter(x, y, color="red", s=50)
# The color parameter specifies the colors of the markers based on the colors array.
 # The s parameter specifies the sizes of the markers based on the sizes array.
plt.show()
````

##**Example 2: Change the size of the dots**
You can change the size of the dots with the s argument.

**Important Note:** Make sure the array for sizes has the same length as the arrays for the x- and y-axis:

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y,  color="red",s=sizes)

plt.show()
````

##**Example 3: Adjust the transparency of the dots**

You can adjust the transparency of the dots with the alpha argument.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes,  color="red", alpha=0.5)

plt.show()
````

##**Example 4: Combine Color Size and Alpha**

- Let's create random arrays with 100 values for x-points, y-points, colors and sizes:

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
````

## **Example 5: Compare two Plots**
The following example generates scatter plots to visualize age distribution across two different age groups: 20-39 years old and 40-69 years old. Each scatter plot represents random data points corresponding to the age group, with markers indicating the distribution of ages. The size of the markers represents the population density within each age group. The code utilizes NumPy to generate random data points and Matplotlib to create the scatter plots. Additionally, comments are provided to explain the purpose of each parameter used in the scatter plot generation.

````python
import numpy as np
import matplotlib.pyplot as plt

AgeGroupOne = range(40, 70, 1)                        # X-values ranging from 40 to 69 year old
Purchase1=np.abs(np.random.randn(30) * 20)             # Y-values generated from a normal distribution

# Scatter plot for ages 40-69
plt.scatter(AgeGroupOne, Purchase1,
    s=200,                                            # Marker size
    alpha=0.6,                                        # Transparency
    label='40-69 years old'                          # Label for the scatter plot
)

plt.scatter(AgeGroupOne,Purchase1, color = '#88c999')

# Scatter plot for ages 20-39
AgeGroupTwo=range(20, 40, 1)    # X-values ranging from 20 to 39 years old
Purchase2=np.abs(np.random.randn(20) * 40)    # Y-values generated from a normal distribution
plt.scatter(AgeGroupTwo,Purchase2,
    s=200,                                            # Marker size
    alpha=0.6,                                        # Transparency
    label='20-39 years old'                          # Label for the scatter plot
)
plt.scatter(AgeGroupTwo,Purchase2, color = '#88c984')
# Display the plot
plt.legend()
plt.xlabel("Age")
plt.ylabel("Average purchase per weekend")                                         # Show legend
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt


rng = np.random.RandomState(0) # Creates a random number generator (RNG) object with a specific seed value of 0. This ensures reproducibility of the random numbers generated.
x = rng.randn(100) # Generates an array x of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).
y = rng.randn(100) # Generates an array y of 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).

colors = rng.rand(100) #Generates an array colors of 100 random numbers between 0 and 1.
sizes = 1000 * rng.rand(100) #Generates an array sizes of 100 random numbers between 0 and 1000.

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.title("Scatter chart")
plt.xlabel("Random data for x")
plt.ylabel("Random data for y")

 # Creates a scatter plot with x values on the x-axis, y values on the y-axis.
 # The c parameter specifies the colors of the markers based on the colors array.
 # The s parameter specifies the sizes of the markers based on the sizes array.
 # The alpha parameter controls the transparency of the markers.
 # The cmap parameter specifies the colormap to use for mapping the colors array values to colors.
plt.colorbar()
plt.show()
````

**Notice** that the color argument is automatically mapped to a color scale (shown here by the** colorbar()** command), and the size argument is given in pixels. In this way, the color and size of points can be used to convey information in the visualization, in order to illustrate multidimensional data.

##**Example 7: Real Time Data analysis and Data Visualization**

Now letâ€™s analyse on a realtime data. So in the code below, I will use the housing prices and I will plot the data to analyze the median prices and the population. The idea is to visualize the densely populated areas with bigger circles and the areas with high prices with darker circles and vice versa:

````python
import pandas as pd
housing_df = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")
housing_df
````

````python
plt.figure(figsize=(12, 8))  # Create a new figure with specified size
plt.scatter(
    x=housing_df['longitude'],  # x-values
    y=housing_df['latitude'],   # y-values
    alpha=0.4,                  # Transparency of markers
    s=housing_df['population'] / 100,  # Marker sizes based on population
    c=housing_df['median_house_value'],  # Color values based on median_house_value
    cmap=plt.get_cmap('jet')    # Colormap
)

#c='median_house_value' specifies the color of the markers based on the 'median_house_value' column. However, this parameter expects a sequence of color values or a colormap, not a column name.
plt.colorbar(label='Median House Value')  # Add colorbar with label
plt.legend(['Population'])  # Add legend with label
plt.xlabel('Longitude')     # x-axis label
plt.ylabel('Latitude')      # y-axis label
plt.title('Housing Data Scatter Plot')  # Plot title
plt.show()  # Show the plot
````
