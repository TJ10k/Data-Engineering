##**Example 1 - First Waffle Chart**

The waffle chart can now be created using the Waffle class and the figure method. With the rows and columns arguments you can specify the number of rows and columns you want your waffle to have. By default, the number of squares in the grid adapts to your number of rows/columns.

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

# create simple dummy data
data = {'Cat': 15, 'Dog': 30, 'hippo': 10, 'Fish': 20, 'horse': 25}


# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns =10,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    colors=["#FF5733", "#FFC300", "#DAF7A6", "#5DADE2", "#9B59B6"]
)


plt.title('Pet Distribution')

plt.show()
```

##**Example 2: Color Customizations**
**colors:** The colors argument changes the color of the grids and must be a list of colors of the same length of your number of categories.


```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

# create simple dummy data
data = {'Cat': 15, 'Dog': 30, 'hippo': 10, 'Fish': 20, 'horse': 25}

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
   colors=["#EFE300", "#FFC300", "#900C3F", "#FF5733", "#DAF7A6"], # Replace with your colors
  # You can use different icons like 'circle', 'diamond', 'square', 'hexagon', etc.
     figsize=(8, 4),  # Adjust figsize to change the width of the waffle chart
    font_size = 60,   # Size of the grids
    icon_legend=True
)

plt.title('Pet Distribution')

plt.show()
```

##**Example 3 - Pictograms: characters and icons**
The pywaffle library can also be used to create pictograms; this is waffle charts, but with characters or icons instead of the default rectangles used to represent the different groups. The characters argument takes Unicode characters as input, such as '♥', '★' or '⬤'.

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

# create simple dummy data
data = {'Cat': 15, 'Dog': 30, 'hippo': 10, 'Fish': 20, 'horse': 25}

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    colors=["#EFE300", "#FFC300", "#900C3F", "#FF5733", "#DAF7A6"], # Replace with your colors
    characters = '♥',  # You can use different icons like 'circle', 'diamond', 'square', 'hexagon', etc.
   figsize=(8, 4),  # Adjust figsize to change the width of the waffle chart
    font_size = 20,   # Size of the grids
    icon_legend=True
)

plt.title('Pet Distribution')

plt.show()
```

##**Example 4 -FontAwesome icons**
