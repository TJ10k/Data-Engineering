## 6. validate_user(email, password) Method:

➢ Establish a connection to the RegistrationDB.

➢ Return True if a user exists with the given email and password. Return False otherwise.


```python
import mysql.connector

class UserTableManipulation:
    def validate_user(self, email, password):
        try:
            # Establish connection to the RegistrationDB
            connection = mysql.connector.connect(
                host="localhost",
                user="root",  # replace with your MySQL username
                password="password",  # replace with your MySQL password
                database="RegistrationDB",
                port="3306"
            )
            cursor = connection.cursor()
            print("Connection established to RegistrationDB")

            # Check if a user exists with the given email and password
            try:
                cursor.execute("SELECT * FROM user WHERE email = %s AND password = %s", (email, password))
                user = cursor.fetchone()
                if user:
                    return True
                else:
                    return False
            except mysql.connector.Error as err:
                print("Error:", err)

            # Close connection
            try:
                cursor.close()
                connection.close()
                print("Connection closed")
            except mysql.connector.Error as err:
                print("Error:", err)

        except mysql.connector.Error as err:
            print("Error:", err)

if __name__ == "__main__":
    user_manipulation = UserTableManipulation()
    email = input("Enter email: ")
    password = input("Enter password: ")
    if user_manipulation.validate_user(email, password):
        print("User exists with the provided email and password")
    else:
        print("No user found with the provided email and password")
```
