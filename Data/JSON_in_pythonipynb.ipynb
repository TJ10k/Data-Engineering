# JSON File Handling Using Python

```python
import json
json_path = 'Data-Engineering\Data\data.json'
with open(json_path) as json_file:
# deseralize data from the file into python dictionary object
    data = json.load(json_file)
    print(data)
```

```python
import json 
my_object = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# convert to JSON
json_string = json.dumps(my_object)
print(json_string)
```
