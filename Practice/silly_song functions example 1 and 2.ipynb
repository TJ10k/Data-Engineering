```python
#Example 1
# Save cell as  silly_song_alltogether.py


# Define a function to sing the refrain
def sing_refrain():
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark!")

# Define a function to sing a verse of our silly song
def sing_verse(lineage):
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark!")

# Define the main function
def main():
    print("Let's sing a silly song!")
    sing_verse("Mommy")   # Call the sing_verse function with "Mommy" as the lineage
    sing_verse("Daddy")  # Call the sing_verse function with "Daddy" as the lineage
    sing_verse("Grandma") # Call the sing_verse function with "Grandma" as the lineage
    sing_verse("Grandpa") # Call the sing_verse function with "Grandma" as the lineage
    sing_refrain()        # Call the sing_refrain function

# Check if this script is being run as the main program
#see val's file called if name equals main for refresher on the next 2 lines:
if __name__ == "__main__":
    main()

""" #Explanation
#reference: https://youtu.be/XqZsoesa55w
# This example is a playful demonstration of organizing code into functions, function definitions, function calls, and script execution.

1. We create a Python script named `silly_song.py`.

2. Inside the script, we define two functions:
   - `sing_refrain()`: This function prints the refrain of a silly pop culture song.
   - `sing_verse(lineage)`: This function prints a verse of the same song with the specified lineage name.

3. We also define the `main()` function, which serves as the entry point of our program.

4. Inside the `main()` function, we print a message to indicate that we are about to sing a silly song.

5. We call the `sing_verse()` function three times, each with a different lineage name ("Mommy," "Daddy," and "Grandma") to create humorous variations of the song.

6. Finally, we call the `sing_refrain()` function to sing the iconic and silly pop  refrain.


Example extensions:
1. break out the refrain and the main verse into a separate file and using function importing.
See files silly_functions.py and silly_song.py for example of This (example 2 below)
2. See using module function aliases (example 4 below)
3. Input/Output lesson:  prompt for input on who to insert into the sing verse function
    **  to do: 2024.02.07 vlb  make this example
 """
```

```python
##save cell to :  song_functions.py

# Define a function to sing a silly pop culture refrain
def sing_refrain():
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark, doo-doo-doo-doo-doo-doo")
    print("Baby shark!")

# Define a function to sing a verse of our silly song
def sing_verse(lineage):
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark, doo-doo-doo-doo-doo-doo")
    print(f"{lineage} shark!")
```

```python
##save cell to :  silly_song.py
# silly_song.py
import song_functions
#see further below for example using :from song_functions import sing_refrain as sr and from song_functions import sing_verse as sv

# Define the main function
def main():
    print("Let's sing a silly song!")
    song_functions.sing_verse("Mommy")   # Call the sing_verse function with "Mommy" as the animal
    song_functions.sing_verse("Daddy")  # Call the sing_verse function with "Daddy" as the animal
    song_functions.sing_verse("Grandma") # Call the sing_verse function with "Grandma" as the animal
    song_functions.sing_verse("Grandpa") # Call the sing_verse function with "Grandma" as the animal
    song_functions.sing_refrain()        # Call the sing_refrain function


if __name__ == "__main__":
    main()  # Call the main function using if __name__ == "__main__"  Remember, a function wont run unless you actually call it. and using if name == main ensures that the correct function is specified as the start of the program.

```
