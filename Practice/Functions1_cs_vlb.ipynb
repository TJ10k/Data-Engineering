```python
# Study Guide: Python Creating Functions

# Explanation: Functions in Python are blocks of reusable code that perform specific tasks. They are defined using the 'def' keyword, followed by the function name and a set of parentheses containing optional parameters. Functions can have a return statement to send back a value as the result of the function.

# Basic Syntax of a Function:
# def function_name(parameters):
#     # Function's block of code
#     return result

# Example 1: Creating a Simple Function
def greet(name):
    return f"Hello, {name}!"

# Use Case: The 'greet' function can be used to greet users with personalized messages by providing their names as arguments.

# How to Call the Function:
result = greet("Alice")
print(result)

# Explanation: In Example 1, we define a function named 'greet' that takes a 'name' parameter. The function returns a greeting message using the provided name.

# Explanation: We call the 'greet' function with the argument "Alice". The function is executed with the provided argument, and the returned value is stored in the 'result' variable, which is then printed.

# Real-world Example 1: Calculating Total Price
def calculate_total_price(price, quantity):
    """This function calculates the total price based on the given price and quantity."""
    total = price * quantity
    return total

# Use Case: The 'calculate_total_price' function can be used to determine the total cost of purchasing items by providing the price per item and the quantity as arguments.

# How to Call the Function:


#conceptually this is how we explain the call below:
# # for each item in order, extract item id, and quantity.
# for each item id, reference the inventory spreadsheet.... match the item number for each row in inventory cost spreadsheet, extract the price per item.


price_per_item = 10.99
quantity = 5
total_amount = calculate_total_price(price_per_item, quantity)
print(f"The total amount for {quantity} items is ${total_amount:.2f}.")

# Explanation: In this real-world example for calculate_total_price, we call the function with 'price_per_item' as 10.99 and 'quantity' as 5. The function calculates the total amount and returns it, which is then printed with two decimal places using the f-string formatting.

# Real-world Example 2: Checking Eligibility
def check_eligibility(age):
    # define a function name'check_eligibility' takes 'age' as a parameter.
    if age >= 18:
        return "You are eligible to vote."
    else:
        return "You are not eligible to vote."

# Use Case: The 'check_eligibility' function can be used to determine if a person is eligible to vote based on their age.

# How to Call the Function:
person_age = 22
eligibility_status = check_eligibility(person_age)
print(eligibility_status)

# Explanation: In this real-world example for check_eligibility, we call the function with 'person_age' as 22. The function checks if the person's age is 18 or above and returns an eligibility status message, which is then printed.

# Conclusion:
# Python functions allow us to organize our code, make it more modular, and improve its reusability. By creating functions, we can encapsulate specific tasks into separate blocks and call them whenever needed. Understanding how to create and call functions is essential for writing efficient and well-structured Python programs. The study guide provides basic syntax, use cases, and examples to illustrate these concepts in action.

```
